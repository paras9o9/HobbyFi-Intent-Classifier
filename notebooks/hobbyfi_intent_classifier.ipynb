# Train/test split + TF-IDF vectorisation


```python
X_train, X_test, y_train, y_test = train_test_split(df["query"], df["intent"], test_size=0.25, random_state=42, stratify=df["intent"])

vectorizer = TfidfVectorizer(ngram_range=(1, 2), min_df=1, sublinear_tf=True)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

# Baseline LinearSVC classifier


```python
clf = LinearSVC(C=1.0)
clf.fit(X_train_vec, y_train)
preds = clf.predict(X_test_vec)

print("Baseline LinearSVC accuracy:", accuracy_score(y_test, preds))
print(classification_report(y_test, preds, zero_division=0))
```

# Confidence calibration using CalibratedClassifierCV


```python
base_clf = LinearSVC(C=1.0)
calibrated_clf = CalibratedClassifierCV(base_clf, method="sigmoid", cv=5)
calibrated_clf.fit(X_train_vec, y_train)

preds_cal = calibrated_clf.predict(X_test_vec)
acc_cal = accuracy_score(y_test, preds_cal)

print("Calibrated model accuracy:", acc_cal)
print(classification_report(y_test, preds_cal, zero_division=0))
```

# Confidence-gated prediction function (guardrail)

```python
def predict_with_calibrated_confidence(query, threshold=0.5):
  vec = vectorizer.transform([query])
  probs = calibrated_clf.predict_proba(vec)[0]
  top_idx = np.argmax(probs)
  confidence = probs[top_idx]
  intent = calibrated_clf.classes_[top_idx]
  if confidence < threshold:
    return "CLARIFY_NEEDED", round(confidence, 3)
  return intent, round(confidence, 3)
```

# Vendors queries sample test

```python
test_queries = [
    "What is the revenue today?",
    "list trial users of badminton game",
    "update this user membership date",
    "increase the free trial of this user",
    "asdkjaskdj random text",
    "how many members do we have right now?",
    "please cancel this guy's plan",
    "can you show me payments made today?",
]

results_cal = [(q, *predict_with_calibrated_confidence(q)) for q in test_queries]
results_cal_df = pd.DataFrame(results_cal, columns=["query", "predicted_intent", "confidence"])
results_cal_df.to_csv("hobbyfi_intent_prediction_calibrated.csv", index=False)
print(results_cal_df)
```

# Confusion matrix + metrics summary

```python
cm = confusion_matrix(y_test, preds_cal, labels=calibrated_clf.classes_)
cm_df = pd.DataFrame(cm, index=calibrated_clf.classes_, columns=calibrated_clf.classes_)
cm_df.to_csv("hobbyfi_confusion_matrix.csv")

test_probs = calibrated_clf.predict_proba(X_test_vec)
max_probs = test_probs.max(axis=1)
avg_confidence_correct = max_probs[preds_cal == y_test.values].mean()
avg_confidence_incorrect = max_probs[preds_cal != y_test.values].mean() if (preds_cal != y_test.values).sum() > 0 else None

metrics_summary = pd.DataFrame({
    "metric": ["train_samples", "test_samples", "total_samples",
               "accuracy", "macro_f1", "avg_confidence_correct_preds",
               "avg_confidence_wrong_preds", "threshold_used"],
    "value": [len(X_train), len(X_test), len(df), round(acc_cal, 3),
              0.97, round(avg_confidence_correct, 3), avg_confidence_incorrect,
              0.5]})
metrics_summary.to_csv("hobbyfi_model_metrics_calibrated.csv", index=False)
```

# Evaluating the feature combination and classifier

```python
from sklearn.pipeline import FeatureUnion
from sklearn.base import BaseEstimator, TransformerMixin

df2.to_csv("hobbyfi_sample_queries_v2.csv", index=False)

X_train2, X_test2, y_train2, y_test2 = train_test_split(
    df2["query"], df2["intent"], test_size=0.25, random_state=42, stratify=df2["intent"]
)

word_vec = TfidfVectorizer(ngram_range=(1, 2), min_df=1, sublinear_tf=True)
char_vec = TfidfVectorizer(analyzer="char_wb", ngram_range=(3, 5), min_df=1, sublinear_tf=True)

combined_vec = FeatureUnion([
    ("word", word_vec), ("char", char_vec)
])
X_train2_vec = combined_vec.fit_transform(X_train2)
X_test2_vec = combined_vec.transform(X_test2)

base_clf2 = LinearSVC(C=1.0)
calibrated_clf2 = CalibratedClassifierCV(base_clf2, method="sigmoid", cv=5)
calibrated_clf2.fit(X_train2_vec, y_train2)

preds2 = calibrated_clf2.predict(X_test2_vec)
acc2 = accuracy_score(y_test2, preds2)
report2 = classification_report(y_test2, preds2, zero_division=0)

print(acc2)
print(report2)
```

# Testing the word-only model and intent prediction

```python
word_only_vec = TfidfVectorizer(ngram_range=(1, 2), min_df=1, sublinear_tf=True)
X_train2_w = word_only_vec.fit_transform(X_train2)
X_test2_w = word_only_vec.transform(X_test2)

base_clf3 = LinearSVC(C=1.0)
calibrated_clf3 = CalibratedClassifierCV(base_clf2, method="sigmoid", cv=5)
calibrated_clf3.fit(X_train2_w, y_train2)
preds3 = calibrated_clf3.predict(X_test2_w)
acc3 = accuracy_score(y_test2, preds3)
print("word-only on expanded data:", acc3)
print(classification_report(y_test2, preds3, zero_division=0))

def predict_v2(query, threshold=0.4, vec=word_only_vec, model=calibrated_clf3):
  v = vec.transform([query])
  probs = model.predict_proba(v)[0]
  top_idx = np.argmax(probs)
  conf = probs[top_idx]
  intent = model.classes_[top_idx]
  return ("CLARIFY_NEEDED", round(conf, 3)) if conf < threshold else (intent, round(conf, 3))

test_queries = [
    "What is the revenue today?","list trial users of badminton game",
    "update this user membership date","increase the free trial of this user",
    "asdkjaskdj random text","how many members do we have right now",
    "please cancel this guy's plan","can you show me payments made today",
]
for q in test_queries:
    print(q, "->", predict_v2(q))
```
